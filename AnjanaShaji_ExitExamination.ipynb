```python
## Weight has missing value and is object datatype changing it to numberical type
# Extract numeric values and convert to numeric data type
df['weight'] = df['weight'].str.extract('(\d+)').astype(float)
```

```python
df.weight.dtype
```

```python
sns.histplot(df['weight'],kde =True)
plt.show()
```

```python
df['weight']=df['weight'].fillna(df['weight'].mean())
```

```python
df['ball_controlling_skills']=df['ball_controlling_skills'].fillna(df['ball_controlling_skills'].median())
df['jumping_skills']=df['jumping_skills'].fillna(df['jumping_skills'].median())
df['penalties_conversion_rate']=df['penalties_conversion_rate'].fillna(df['penalties_conversion_rate'].median())
df['mental_strength']=df['mental_strength'].fillna(df['mental_strength'].mean())
df['shot_accuracy']=df['shot_accuracy'].fillna(df['shot_accuracy'].median())
df['behaviour_rating']=df['behaviour_rating'].fillna(df['behaviour_rating'].median())
df['fitness_rating']=df['fitness_rating'].fillna(df['fitness_rating'].median())
df['years_of_experience']=df['years_of_experience'].fillna(df['years_of_experience'].median())
```

```python
df['gender'].mode(),df['strong_foot'].mode(),df['coaching'].mode()
```

```python
df['gender']=df['gender'].fillna(df['gender'].mode()[0])
df['strong_foot']=df['strong_foot'].fillna(df['strong_foot'].mode()[0])
df['coaching']=df['coaching'].fillna(df['coaching'].mode()[0])
```

```python
df['matches_played']=df['matches_played'].fillna(df['matches_played'].median())
```

# Encoding

```python
df.info()
```

```python
# Replace the single quote with a period and convert to numeric
df['height'] = df['height'].str.replace("'", '.').str.replace('"', '').astype(float)
```

```python
df.head()
```

```python
df['dedication_level'].unique()
```

```python
##ordinal encoding dedication level column
from sklearn.preprocessing import OrdinalEncoder


dedication = ['Low',"Medium","High"]

enc = OrdinalEncoder(categories=[dedication])

df[['dedication_level']] = enc.fit_transform(df[['dedication_level']])
```

```python
##onehotencoding on gender and recruitmentchannel Department
from numpy import int32
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False,dtype=int32,drop='first')
cols = df[['gender','strong_foot','coaching']]
onehot_encoded = ohe.fit_transform(cols)
```

```python
feature_names = ohe.get_feature_names_out(['gender','strong_foot','coaching'])
# Create a new DataFrame with the encoded data and feature names
df_encoded = pd.DataFrame(onehot_encoded, columns=feature_names)
# Concatenate the original DataFrame and the encoded DataFrame
df = pd.concat([df, df_encoded], axis=1)
```

```python
df.drop(['gender','strong_foot','coaching'],axis=1,inplace=True)
```

```python
cat_data = [features for features in df.columns if df[features].dtypes == 'O']
cat_data
```

# Spliting Data

```python
## spliting dependent and independent variable
x=df.drop('selection',axis=1)
y=df['selection']
```

```python
y.value_counts(normalize=True)  # data is no high imbalance between data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=123)
```

# Training and Testing Model

```python
from sklearn.metrics import confusion_matrix , accuracy_score, precision_score, recall_score, f1_score,classification_report
```

```python
# 1. Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr_model=lr.fit(x_train,y_train)
lr_pred=lr_model.predict(x_test)


# 2. SVM
from sklearn.svm import SVC
svc = SVC()
svc_model=svc.fit(x_train,y_train)
svm_pred = svc_model.predict(x_test)

#3. Desicion Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt_model=dt.fit(x_train,y_train)
dt_pred=dt_model.predict(x_test)


# 4.Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf_model=rf.fit(x_train,y_train)
rf_pred=rf_model.predict(x_test)

from xgboost import XGBClassifier
xg = XGBClassifier()
xg_model = xg.fit(x_train,y_train)
xg_pred=xg_model.predict(x_test)


## Accuracy

print('Logistic Regression Accuracy is : ', accuracy_score(y_test, lr_pred))
print('SVC Model Accuracy is: ', accuracy_score(y_test, svm_pred))
print('Decision Tree Accuracy is: ', accuracy_score(y_test, dt_pred))
print('Random FOrest Accuracy is: ', accuracy_score(y_test, rf_pred))
print('XG boost Accuracy is: ', accuracy_score(y_test, xg_pred))
```

```python
print('Logistic Regression Report is : \n', classification_report(y_test, lr_pred))
print('SVC Model report is: \n', classification_report(y_test, svm_pred))
print('Decision Tree Report is: \n', classification_report(y_test, dt_pred))
print('Random FOrest Report is: \n', classification_report(y_test, rf_pred))
print('XG boost Report is: \n', classification_report(y_test, xg_pred))
```

# Hypertuning

```python
n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]
max_features = ['auto','sqrt','log2']
max_depth = [int(x) for x in np.linspace(start=10, stop=1000, num=10)]
min_samples_split = [2,5,10,14]
min_samples_leaf = [1,2,4,6,8]
```

```python
random_grid = {'n_estimators':n_estimators,
               'max_features' :max_features,
               'max_depth' : max_depth,
               'min_samples_split' : min_samples_split,
               'min_samples_leaf':min_samples_leaf,
               'criterion': ['entropy','gini','log_loss']}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_randomcv = RandomizedSearchCV(estimator = rf,param_distributions=random_grid,scoring='accuracy',
                                 n_iter = 100, cv=3, verbose=2,random_state=100, n_jobs=-1)
rf_randomcv.fit(x_train,y_train)
```

```python
rf_randomcv.best_params_
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'criterion': [rf_randomcv.best_params_['criterion']],
    'max_depth': [rf_randomcv.best_params_['max_depth']],
    'max_features': [rf_randomcv.best_params_['max_features']],
    'min_samples_leaf': [rf_randomcv.best_params_['min_samples_leaf'],
                         rf_randomcv.best_params_['min_samples_leaf']+2,
                         rf_randomcv.best_params_['min_samples_leaf'] + 4],
    'min_samples_split': [rf_randomcv.best_params_['min_samples_split'] - 2,
                          rf_randomcv.best_params_['min_samples_split'] - 1,
                          rf_randomcv.best_params_['min_samples_split'],
                          rf_randomcv.best_params_['min_samples_split'] +1,
                          rf_randomcv.best_params_['min_samples_split'] + 2],
    'n_estimators': [ rf_randomcv.best_params_['n_estimators'] - 100,
                     rf_randomcv.best_params_['n_estimators'],
                     rf_randomcv.best_params_['n_estimators'] + 100]
}

print(param_grid)
```

```python
grid_search = GridSearchCV(estimator=rf,param_grid = param_grid,cv=10,n_jobs=-1,verbose=2)
grid_search.fit(x_train,y_train)
```

```python
best_params = grid_search.best_params_
best_param_model = grid_search.best_estimator_
```
